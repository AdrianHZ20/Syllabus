## Introducción
En esta ayudantía trabajaremos el manejo y levantamiento de excepciones en Python. El objetivo es comprender cómo las excepciones permiten detectar y controlar errores en el tiempo de ejecución de un programa sin que este se detenga abruptamente. Trabajaremos diferentes tipos de excepciones y practicaremos el uso del bloque `try` (`except`, `else` y `finally`).

Primero, veremos unos ejemplos que nos permiten entender el funcionamiento del bloque `try/except`:

```python
a = int(input())
b = int(input())

try:
    print(f"División: {a/b}")
    print("Dividí correctamente")

except ZeroDivisionError:
    print("Ups! querías dividir por cero")

else:
    print("Todo bien :p")
```

También podemos especificar bloques `except` para distintos tipos de errores:

```python
a = int(input())
b = int(input())

lista = ["cero", "uno", "dos", "tres"]
diccionario = {0: "cero", 1: "uno", 2: "dos", 3: "tres"}

try:
    print(f"División: {a/b}")
    print("Dividí correctamente")
    print(f"Elemento: {lista[a]}")  # prueba cambiando lo que está dentro por diccionario[a]

except ZeroDivisionError:
    print("Ups! querías dividir por cero")

except (IndexError, KeyError):
    print("Ups! querías acceder a un elemento que no existe")

else:
    print("Todo bien :p")
```

El bloque `finally` **siempre se ejecuta**, aunque haya un error que no se capture.

```python
a = int(input())
b = int(input())

lista = ["cero", "uno", "dos", "tres"]
diccionario = {0: "cero", 1: "uno", 2: "dos", 3: "tres"}

try:
    print(f"División: {a/b}")
    print("Dividí correctamente")
    print(f"Elemento: {lista[a]}")  # prueba cambiando lo que está dentro por diccionario[a]
    lista.remove("cuatro")  # esto nos arroja un ValueError

except ZeroDivisionError:
    print("Ups! querías dividir por cero")

except (IndexError, KeyError):
    print("Ups! querías acceder a un elemento que no existe")

else:
    print("Todo bien :p")

finally:
    print("Fin del código")
```

## Ejercicio 1: ¡Elaboremos un mapa! 🗺️

El pirata **Beniju** se embarca en una aventura para explorar distintas islas y necesita **construir sus mapas** para recordar lo que hay en cada una.  

Cada **mapa** es una grilla rectangular con celdas donde se pueden colocar elementos como **"Cofre"**, **"Tesoro"**, **"Roca"** o **"Mar"**. El pirata también debe guardar todos los mapas que cree en su bitácora personal.  

Tu tarea es **implementar los siguientes métodos** que permitan al pirata crear y completar sus mapas:  

### 1. `Mapa.agregar_elemento(x, y, elemento)`  

- Recibe las coordenadas `x, y` y un `elemento` para ubicar en la celda correspondiente.
- Dispones del atributo de Mapa **celdas**, que es una _lista de listas_ que comienza inicialmente vacía y se debe rellenar con el elemento recibido en la coordenada indicada.
- Si las coordenadas están **fuera de los límites** del mapa, se debe **levantar la excepción `CoordenadaInvalida(x, y)`**.  
- Si la celda ya tiene un elemento, se debe **levantar la excepción `CoordenadaOcupada(x, y, elemento_existente)`**, siendo elemento_existente el elemento que ocupa la celda.  
- Si no ocurrió lo anterior y por ende se pudo agregar el elemento, se debe imprimir `Elemento {elemento} agregado con éxito a {nombre_mapa} en {x}, {y}`

### 2. `Pirata.crear_mapa(nombre_mapa, ancho, alto, elementos_mapa)`  

- Crea una instancia de `Mapa` con el `nombre_mapa` y las dimensiones (`ancho`, `alto`).  
- Recibe una lista `elementos_mapa`, que contiene tuplas de la forma `(x, y, elemento)` y se debe intentar agregar cada elemento en la posición indicada usando `agregar_elemento`.  
- Si ocurre una excepción (`CoordenadaInvalida` o `CoordenadaOcupada`), **debes capturarla e imprimir su mensaje**, sin detener la creación del resto de elementos.  
- Una vez terminado, el nuevo mapa debe ser **añadido a la lista de mapas del pirata (`self.mapas`)** y debe mostrarse en pantalla usando la función `imprimir_mapa(mapa.celdas)` del módulo utils.

Sugerencia: para ver cómo funcionan las excepciones personalizadas, puedes correr el código **sin capturar la excepción en la función anterior.**

## Ejercicio 2: ¡A saquear cofres! 💎

Beniju ya puede crear mapas y rellenarlos con elementos. Ahora debe leer sus mapas para poder **abrir cofres** y **obtener botín** (o sufrir trampas). Para ello debes implementar los siguientes métodos:

### 1. `Pirata.obtener_loot(mapa, x, y)`  
Este método recibe un objeto `Mapa` y unas coordenadas. Su tarea es retornar el contenido del cofre o levantar excepciones en caso de trampas.

- Si en las coordenadas `(x, y)` del mapa no hay un **`'Cofre'`**, se debe levantar la excepción `NoEsCofre(x, y, mapa)`.  
- Si efectivamente hay un cofre, primero se elige un botín al azar de la lista `loot_cofres` del módulo utils.
- Se debe consultar la probabilidad de que se entregue el botín, buscando el valor asociado al nombre del botín en `probabilidad_loot`. Si el botín no se encuentra allí, usa un valor por defecto de `0.5`. (Las probabilidades que entregue `probabilidad_loot` siempre estarán entre 0 y 1)
- Con esa probabilidad decides:
  - Si ocurre el evento “favorable”, se devuelve el botín elegido.  
  - En caso contrario, el cofre estaba “trucado” y se debe levantar **una excepción al azar**, escogiendo entre `CofreBomba`, `CofreVeneno` o `CofreMaldicion`.  

### 2. `Pirata.abrir_cofre(self, x, y)`  
Este método recibe unas coordenadas e intenta abrir cofres en **todos los mapas** que tenga registrados el pirata.

- Se debe validar que `x` e `y` sean enteros; de lo contrario, debe levantar un `TypeError` con el siguiente mensaje `⚠️ERROR: Las coordenadas deben ser valores enteros`.  
- Luego para cada mapa del pirata debe:
  - Verificar que las coordenadas no estén fuera de los límites del mapa, en ese caso, se debe imprimir `⚠️ERROR: La coordenada {x}, {y} se encuentra fuera del mapa: {nombre_mapa}`
  - Si en la celda hay un `'Cofre'`, se debe llamar a `obtener_loot`:
    - Si retorna un botín, se agrega a `self.objetos_conseguidos`.
    - Si se levanta alguna de las trampas (`CofreBomba`, `CofreVeneno`, `CofreMaldicion`), se debe imprimir el siguiente mensaje: `Has esquivado un {tipo_cofre} de {nombre_mapa} en {x}, {y}` (Siendo los tipos de cofre el nombre de la excepción obtenida)
- Independientemente del resultado, la celda donde estaba el cofre debe quedar vacía al final del proceso. (Recordar que una celda vacía está descrita por None)

---

```python
from exceptions import CoordenadaOcupada, CoordenadaInvalida, CofreBomba, CofreVeneno, CofreMaldicion, NoEsCofre
from utils import imprimir_mapa, loot_cofres, probabilidad_loot
import random

class Mapa:
    def __init__(self, nombre: str, ancho: int, alto: int):
        self.nombre = nombre
        self.celdas = [[None for _ in range(ancho)] for _ in range(alto)]

    def agregar_elemento(self, x, y, elemento):
        # TODO:
        try:
            if self.celdas[y][x]:
                raise CoordenadaOcupada(x, y, self.celdas[y][x])
        except IndexError:
            raise CoordenadaInvalida(x, y)
        else:
            self.celdas[y][x] = elemento

        print(f'Elemento {elemento} agregado con éxito a {self.nombre} en {x}, {y}')

class Pirata:
    def __init__(self, nombre):
        self.nombre = nombre
        self.mapas: list[Mapa] = []
        self.objetos_conseguidos = []
    
    def crear_mapa(self, nombre_mapa, ancho, alto, elementos_mapa):
        # TODO:
        mapa = Mapa(nombre_mapa, ancho, alto)
        for elemento in elementos_mapa:
            try:
                mapa.agregar_elemento(*elemento)
            except (CoordenadaInvalida, CoordenadaOcupada) as e:
                print(e)
        self.mapas.append(mapa)
        imprimir_mapa(mapa.celdas)
    
    def obtener_loot(self, mapa: Mapa, x, y):
        # TODO:
        if mapa.celdas[y][x] != 'Cofre':
            raise NoEsCofre(x, y, mapa)
        
        loot = random.choice(loot_cofres)
        try:
            cofre_sin_trampa = probabilidad_loot[loot]
        except KeyError:
            cofre_sin_trampa = 0.5

        if random.random() <= cofre_sin_trampa:
            return loot
        else:
            raise random.choice([CofreBomba(), CofreVeneno(), CofreMaldicion()])

    def abrir_cofre(self, x, y):
        # TODO:
        if not isinstance(x, int) or not isinstance(y, int):
            raise TypeError(f"⚠️ERROR: Las coordenadas deben ser valores enteros")
        
        for mapa in self.mapas:
            try:
                if mapa.celdas[y][x] == 'Cofre':
                    try:
                        self.objetos_conseguidos.append(self.obtener_loot(mapa, x, y))
                    except CofreBomba:
                        print(f'Has esquivado un CofreBomba de {mapa.nombre} en {x}, {y}')
                    except CofreVeneno:
                        print(f'Has esquivado un CofreVeneno de {mapa.nombre} en {x}, {y}')
                    except CofreMaldicion:
                        print(f'Has esquivado un cofre CofreMaldicion de {mapa.nombre} en {x}, {y}')
                    finally:
                        mapa.celdas[y][x] = None
            except IndexError:
                print(f'⚠️ERROR: La coordenada {x}, {y} se encuentra fuera del mapa: {mapa.nombre}')
            

if __name__ == "__main__":
    random.seed('DCCazador de Tesoros')

    pirata = Pirata('Beniju')

    elementos_mapa_1 = [
        (1, 3, "Cofre"), (3, 4, "Roca"), (10, 5, "Cofre"),
        (2, 2, "Cofre"), (1, 2, "Tesoro"), (4, 4, "Cofre"),
        (6, 7, "Mar"), (5, 2, "Cofre"), (0, 6, "Cofre"),
        (6, 0, "Cofre"), (3, 0, "Roca"), (1, 3, "Tesoro"),
        (4, 1, "Mar"), (2, 2, "Tesoro")
    ]
    elementos_mapa_2 = [
        (0, 0, "Cofre"), (5, 4, "Roca"), (1, 1, "Cofre"),
        (2, 1, "Mar"), (0, 4, "Tesoro"), (3, 0, "Cofre"),
        (2, 2, "Roca"), (4, 4, "Cofre"), (4, 5, "Mar"),
        (1, 1, "Tesoro"), (0, 0, "Roca")
    ]
    elementos_mapa_3 = [
        (0, 2, "Cofre"), (6, 1, "Roca"), (2, 0, "Cofre"),
        (4, 1, "Mar"), (1, 4, "Cofre"), (3, 3, "Cofre"),
        (3, 2, "Tesoro"), (2, 6, "Mar"), (5, 5, "Cofre"),
        (0, 5, "Roca"), (3, 3, "Tesoro"), (0, 2, "Mar")
    ]
    elementos_mapa_4 = [
        (7, 0, "Cofre"), (8, 2, "Tesoro"), (1, 0, "Roca"),
        (0, 5, "Cofre"), (4, 1, "Cofre"), (3, 5, "Mar"),
        (6, 4, "Cofre"), (2, 2, "Cofre"), (7, 6, "Mar"),
        (7, 0, "Tesoro"), (2, 2, "Roca"), (5, 3, "Tesoro")
    ]

    pirata.crear_mapa("Isla Misteriosa", 7, 7, elementos_mapa_1)
    pirata.crear_mapa("Bahía Oscura", 5, 5, elementos_mapa_2)
    pirata.crear_mapa("Cayo Calavera", 6, 6, elementos_mapa_3)
    pirata.crear_mapa("Atolón Fantasma", 8, 6, elementos_mapa_4)

    print("\n--- Aperturas de cofres ---")
    aperturas = [
        (1, 3), (2, 2), (4, 4), (5, 2), (0, 6), (6, 0),
        (0, 0), (1, 1), (3, 0), (4, 4), (2, 3),
        (0, 2), (2, 0), (1, 4), (5, 5), (3, 3),
        (7, 0), (0, 5), (4, 1), (6, 4), (2, 2)
    ]
    for x, y in aperturas:
        print(f"\nIntentando abrir cofre en ({x}, {y})...")
        pirata.abrir_cofre(x, y)

    print("\n--- Caso fuera de rango ---")
    pirata.abrir_cofre(99, 99)

    print("\n--- Caso tipos inválidos ---")
    try:
        pirata.abrir_cofre("2", 3.0)
    except TypeError as e:
        print(f"TypeError capturado correctamente: {e}")

    print("\n--- Resumen ---")
    print("Objetos conseguidos:", pirata.objetos_conseguidos)

```
